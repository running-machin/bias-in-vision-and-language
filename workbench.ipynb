```python
import os
import requests
import concurrent.futures
import subprocess
# import logging
# Set up logging configuration
# logging.basicConfig(filename='logs.txt', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
def download_image(item, folder_name, subfolder_name):
    url = item['url']
    try:
        # response = requests.get(url, allow_redirects=True)
        # if response.status_code != 200:
        #     print(url, item['image_name'])
        #     item['active'] = False
        # else:
        #     item['active'] = True
        image_path = os.path.join(os.getcwd(), 'data', 'google-images', folder_name, subfolder_name, item['image_name'])
        if not os.path.exists(image_path):
            subprocess.run('torsocks wget -O {} {}'.format(image_path, url), shell=True)
        # print(image_path)
        if not os.path.exists(image_path):
            item['active'] = False
        else:
            item['active'] = True
            # print(f"{item['image_name']}not donwloaded yet")
            
            
        #     with open(image_path, 'wb') as f:
        #         for chunk in response.iter_content(1024):
        #             f.write(chunk)
        # else:
        #     print(f"{item['image_name']}already downloaded")
    except Exception as e:
        print(f"{e}: for URL: {url} and image_name: {item['image_name']}")
        item['active'] = False

# Use a ThreadPoolExecutor to download images in parallel
with concurrent.futures.ThreadPoolExecutor() as executor:
    for i in range(len(data)):
        folder_name = data[i]['folder_name']
        subfolder_name = data[i]['subfolder_name']
        executor.map(download_image, data[i]['wget_commands'], [folder_name]*len(data[i]['wget_commands']), [subfolder_name]*len(data[i]['wget_commands']))    

# # save the updated json file
# with open('output_2.json', 'w') as f:
#     json.dump(data, f, indent=4)
```

```python
import os
from PIL import Image

def check_and_delete_corrupt_images(image_path):
    try:
        # Try to open the image
        with Image.open(image_path) as img:
            img.verify()
    except FileNotFoundError as e:
        print(f"File not found: {image_path}")
    except (IOError, SyntaxError) as e:
        # If an exception is raised, the image is corrupt
        print(f"Corrupt image: {image_path}, deleting...")
        os.remove(image_path)

counter = 0
# Call the function for each image
for i in range(len(data)):
    for item in data[i]['wget_commands']:
        image_path = os.path.join(os.getcwd(), 'data', 'google-images', data[i]['folder_name'], data[i]['subfolder_name'], item['image_name'])
        check_and_delete_corrupt_images(image_path)
        if not os.path.exists(image_path):
            counter += 1
print(counter)
        # print(image_path)
```

```python
import pandas as pd

df =pd.json_normalize(data)
df = df.explode('wget_commands')
df = df.reset_index()
df_json = pd.json_normalize(df.wget_commands)
df = pd.concat([df, df_json], axis=1).reset_index().drop(columns=['index','level_0','wget_commands'])



def clean_data(df):
    # Derive column 'image_atributes' from column: 'image_name'
    def image_atributes(image_name):
        """
        Transform based on the following examples:
           image_name                   Output
        1: "white_soft_0.jpg"        => "white soft"
        2: "female_1.jpg"            => "female"
        3: "ea_competent_male_0.jpg" => "ea competent male"
        """
        split1 = image_name.split("_")
        if len(image_name) - len(image_name.replace("_", "")) == 2:
            return split1[0] + " " + split1[1]
        if len(image_name) - len(image_name.replace("_", "")) == 1:
            return split1[0]
        if len(image_name) - len(image_name.replace("_", "")) == 3:
            return split1[0] + " " + split1[1] + " " + split1[2]
        return None
    df.insert(3, "image_atributes", df.apply(lambda row : image_atributes(row["image_name"]), axis=1))
    # Replace all instances of "AA" with "African American" in column: 'image_atributes'
    df['image_atributes'] = df['image_atributes'].str.replace("aa ", "African American ", case=False, regex=False)
    # Replace all instances of "ea" with "European American" in column: 'image_atributes'
    df['image_atributes'] = df['image_atributes'].str.replace("ea ", "European American ", case=False, regex=False)
    return df

df = clean_data(df)
print(df.image_atributes.value_counts().head())
```

```python
import requests
import json

host = "http://127.0.0.1:8888"

def text2img(params: dict) -> dict:
    """
    text to image
    """
    result = requests.post(url=f"{host}/v1/generation/text-to-image",
                           data=json.dumps(params),
                           headers={"Content-Type": "application/json"})
    return result.json()

result =text2img({
    "prompt": "1girl sitting on the ground",
    "async_process": True})
print(result)
```
